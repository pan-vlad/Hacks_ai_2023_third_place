####  Предобработка данных

```python
t = pd.DataFrame()
t
```

```python
t['Сервис_без_цифр'] = train['Сервис'].replace('[А-Яа-я-]', '', regex=True)
t['Приоритет_без_цифр'] = train['Приоритет'].replace('[А-Яа-я-]', '', regex=True)
t['Функциональная_группа_без_цифр'] = train['Функциональная группа'].replace('[А-Яа-я-]', '', regex=True)
t['Критичность_без_цифр'] = train['Критичность'].replace('[А-Яа-я-]', '', regex=True)
t['Влияние_без_цифр'] = train['Влияние'].replace('[А-Яа-я-]', '', regex=True)
# t['Место_без_цифр'] = train['Место'].replace('[А-Яа-я-]', '', regex=True)

# t['Место_без_цифр'] = train['Место'].replace('[#/]', '', regex=True)
# t['Место_без_цифр'] = train['Место'].replace('Место', '', regex=True)
# t['Место_без_цифр'] = train['Место'].replace('К', '', regex=True)

t['Тип переклассификации'] = train['Тип переклассификации']
# t['Крайний срок'] = train['Крайний срок']
# t['Дата обращения'] = train['Дата обращения']
# t['Дата восстановления'] = train['Дата восстановления']
# t['Дата закрытия обращения'] = train['Дата закрытия обращения']

t
```

```python
for i in ['Сервис_без_цифр', 'Приоритет_без_цифр', 'Функциональная_группа_без_цифр',
          'Критичность_без_цифр', 'Влияние_без_цифр']:
    t[t[i] == ''] = 0
# t['Влияние_без_цифр'].astype(float)
t[['Сервис_без_цифр', 'Приоритет_без_цифр', 'Функциональная_группа_без_цифр',
          'Критичность_без_цифр', 'Влияние_без_цифр']] = t[['Сервис_без_цифр', 'Приоритет_без_цифр', 'Функциональная_группа_без_цифр',
          'Критичность_без_цифр', 'Влияние_без_цифр']].astype(float)
```

```python
decomposed = seasonal_decompose(t,period=2)

decomposed.trend.plot(ax=plt.gca(), figsize=(12,5))
plt.ylabel('Общий тренд')
plt.title("Весь период")
plt.show()

decomposed.seasonal.plot(ax=plt.gca(), figsize=(12,5))
plt.ylabel('Общая сезонность')
plt.show()

decomposed.resid.plot(ax=plt.gca(), figsize=(12,5))
plt.ylabel('Общий шум')
plt.show()
```
